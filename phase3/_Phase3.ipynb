Seaborn:Seaborn is a Python data visualization library based on Matplotlib, but it's:
Easier to use for statistical graphics,
Prettier and cleaner by default,
Integrated with Pandas DataFrames,
Works directly with DataFrames,
Built-in themes and color palettes,
Plots like heatmaps, box plots, violin plots, pair plots, histograms, and regression plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'math': [88, 75, 92, 68, 80, 95, 60, 85, 70, 78],
    'science': [92, 85, 89, 74, 88, 94, 65, 90, 72, 84],
    'english': [85, 78, 94, 70, 85, 98, 63, 86, 68, 80],
    'history': [70, 65, 90, 60, 85, 92, 58, 88, 65, 77],
    'computer': [95, 80, 98, 72, 90, 99, 66, 91, 75, 85]
}

df = pd.DataFrame(data)
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

Plotly: Plotly is an open-source data visualization library in Python (and other languages) that allows you to create interactive and dynamic charts, graphs, and dashboards â€” perfect for data analysis, reporting, and presentations.

```python
pip install plotly
```

```python
import plotly.express as px
import pandas as pd
data = {
    "Month": ["January", "February", "March", "April", "May"],
    "Revenue": [15000, 18000, 12000, 22000, 25000]
}

df = pd.DataFrame(data)
fig = px.bar(df , x = "Month",y="Revenue",title = "Monthly Revenue", color = "Revenue")
fig.show()
```
